```python
# Convert 'Scheduled Arrival Time' to '%H:%M:%S' format
merged_data['Scheduled Arrival Time'] = pd.to_datetime(merged_data['Scheduled Arrival Time']).dt.strftime('%H:%M:%S')
merged_data['Scheduled departure time'] = pd.to_datetime(merged_data['Scheduled departure time']).dt.strftime('%H:%M:%S')

# Combine 'Date (MM/DD/YYYY)' and 'Scheduled Arrival Time' into a single datetime column
merged_data['Arrival_time'] = pd.to_datetime(merged_data['Date (MM/DD/YYYY)'].astype(str) + ' ' + merged_data['Scheduled Arrival Time'].astype(str))
merged_data['departure_time'] = pd.to_datetime(merged_data['Date (MM/DD/YYYY)'].astype(str) + ' ' + merged_data['Scheduled departure time'].astype(str))
merged_data.head()
```

```python
ord_weather_data = pd.read_csv("ORD_weather_data_sub_hourly.csv")
ord_weather_data['Location'] = 'ORD'
ord_weather_data.head()
```

```python
jfk_weather_data = pd.read_csv("JFK_weather_data_sub_hourly.csv")
jfk_weather_data['Location'] = 'JFK'
jfk_weather_data.head()
```

```python
mco_weather_data = pd.read_csv("MCO_weather_data_sub_hourly.csv")
mco_weather_data['Location'] = 'MCO'
mco_weather_data.head()
```

```python
weather_data = pd.concat([ord_weather_data, jfk_weather_data, mco_weather_data], ignore_index=True)
weather_data.head()
```

```python
# Convert 'timestamp_local' and 'timestamp_utc' to datetime
weather_data['timestamp_local'] = pd.to_datetime(weather_data['timestamp_local'])
weather_data['timestamp_utc'] = pd.to_datetime(weather_data['timestamp_utc'])

# Print the DataFrame to verify the changes
weather_data.head()
```

```python
weather_data.dtypes
```

```python
weather_data['timestamp_local']
```

```python
merged_data['Arrival_time'] = merged_data['Arrival_time'].dt.round('15min')
merged_data.head()
```

```python
merged_data['departure_time'] = merged_data['departure_time'].dt.round('15min')
merged_data.head()
```

```python
merged_data['departure_time']
```

```python
# Convert 'departure_time' to timezone-naive datetime (local time)
merged_data['departure_time'] = merged_data['departure_time'].dt.tz_localize(None)
merged_data['Arrival_time'] = merged_data['Arrival_time'].dt.tz_localize(None)
```

```python

merged_data = pd.merge(merged_data, weather_data, 
                       left_on=['departure_time', 'Origin Airport'], 
                       right_on=['timestamp_local', 'Location'],
                       how='left')
merged_data.head()
```
